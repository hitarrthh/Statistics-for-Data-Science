```python
population_size = 1000
true_proportion_smokers = 0.395
num_samples = 1000
sample_size = 100
population = np.random.choice([0, 1],
size=population_size, p=[1-true_proportion_smokers,
true_proportion_smokers])

def calculate_sample_proportion(sample):
return np.sum(sample) / len(sample)

sample_proportions = np.zeros(num_samples)

for i in range(num_samples):
sample = np.random.choice(population, size=sample_size,
replace=False)
sample_proportions[i] =
calculate_sample_proportion(sample)

plt.hist(sample_proportions, bins=30, density=True,
alpha=0.6, color='b', edgecolor='black')

plt.xlabel('Sample Proportion of Smokers')
plt.ylabel('Density')
plt.title('Sampling Distribution of Sample Proportions')
plt.grid(True)
plt.show()
```
